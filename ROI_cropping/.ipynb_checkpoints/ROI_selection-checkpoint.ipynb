```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

PATH = "D:/MAIA_Masters/IMMAS/dataset/images/"
image_name ="22579730_bbd6a3a35438c11b_MG_R_ML_ANON.tif"
CROP_PATH = "D:/MAIA_Masters/Semester2_Italy/Distributed Programming/Project/Java-Mass-Detection/ROI_cropping/CROPPED/"


im = cv2.imread(PATH.__add__(image_name), 0)
plt.imshow(im,interpolation="nearest",cmap='gray')
plt.show()
```

```python
im.shape
```

```python
def CROP(im, image_name,CROP_KERNEL):
    for i  in range(CROP_KERNEL,im.shape[0],CROP_KERNEL):
        for j  in range(CROP_KERNEL,im.shape[1],CROP_KERNEL):
            cropped_name = image_name+'_'+str(i)+'_'+str(j)+'.png'
            crop = im[i-CROP_KERNEL:i, j-CROP_KERNEL:j]
            if np.mean(crop) <= 3.0:
                continue
            else:
                flag = cv2.imwrite(CROP_PATH.__add__(cropped_name),crop)
                if (flag == True):
                    continue
                else:
                    print("Error in writting cropped image!")
```

```python
CROP_KERNEL = 454
CROP(im, image_name,CROP_KERNEL)
```

```python
import os
IMAGES = os.listdir(PATH)
for i in IMAGES:
    im = cv2.imread(PATH.__add__(i), 0)
    CROP(im, i,CROP_KERNEL)
    
    
```
